```python
#4.Find out under which team and position, spending in terms of salary is high.
```

```python
group=df.groupby(['Team','Position']).sum()
group.reset_index(inplace=True)
heighest_idx=group['Salary'].idxmax()
heighest_spending=group.loc[heighest_idx]
heighest_spending[['Team','Position','Salary']]
```
